```python
def levenshtein_distance(a, b):
    if a == b:
        return 0
    if len(b) >= len(a):
        a, b = b, a
    if not a:
        return len(b)
    
    previous = range(1+ len(b))
    for i, col in enumerate(a):
        current = [i + 1]
        for j, col1 in enumerate(b):
            insertions = previous[j + 1] + 1
            deletions = current[j] + 1
            substitutions = previous[j] + (col != col1)
            current.append(min(insertions, deletions, substitutions))
        previous = current
    # Returns the distance
    return previous[-1]
```

```python
pub_dict = {}
```

```python
for index, row in pub.iterrows():
    nameArr = row['Name'].split(',')
    name = nameArr[0].lower() + nameArr[1][0:2].lower()
    #author_sep = row['Authors'].split(', ')
    if name not in pub_dict:
        pub_dict[name]= [[index, row['Publication type'], row['Title']]]
    else:
        pub_dict[name].append([index, row['Publication type'], row['Title']])

    
```

```python
intel_lst = []
for index, row in intel.iterrows():
    # get all co-authors as LN FIMI, ... 19-79
    startInd = 19
    co_author_str = ""
    for col in row:
        if pd.isna(row.iloc[startInd]):
            break
        if startInd >= 79:
            break
        
        if pd.isna(row.iloc[startInd+2]):
            break
        ln = row.iloc[startInd+2]
        
        if not pd.isna(row.iloc[startInd]):
            fi = row.iloc[startInd][0]
            
        if not pd.isna(row.iloc[startInd+1]):
            mi = row.iloc[startInd+1][0]
        
        co_author_str+= ln +" "+fi+mi+", "
        startInd +=5
    
    #print(row.iloc[startInd])
    #print(pd.isna(row.iloc[79]))
    name = row['Last Name'].lower()+" "+row['First Name'][0].lower()
    lst = [index, name, row['CONTYPE'],  row['TITLE'], co_author_str[:-2]]
    intel_lst.append(lst)
    
```
