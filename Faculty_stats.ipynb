```python
# Return stat if exists, else return an empty list
def get_stats(a, author_stats):   
    if a in author_stats:
        stat_a = author_stats[a]
    else:
        stat_a = [0,0,0] #,set(),set(),set()
    
    return stat_a
```

Logic below - 
For each publication, a dictionary for college and deaprtments is created as <dept/col name, count>
If count is more than 2, then it means 2 or more authots have same deap/ col.
In these cases increase the deptartment count and college count for the authors

```python
author_stats = {}
count = 0

for index, row in pub_sample.iterrows():
    #print(index)
    count +=1
    try:
        nameArr = row['Name'].split(',')
        name = nameArr[0].lower() + nameArr[1][0:2].lower()
        author_sep = row['Authors'].split(', ') # , and space
    except:
        print(str(index)+" "+row['Name'])
        continue
    
    co_name = []
    for co_a in author_sep:
        
        try:
            co_name.append(co_a.split(' ')[0].lower()+" "+co_a.split(' ')[1][0:1].lower()) #Last name + space + first name first char
        except:
            continue
    
    col_map = {}
    dept_map = {}
    added = False
    #print(name)
    print(co_name)            

    # create a map of dept name, college name and counts for all the authors
    for name_co in co_name:
        
        if name_co in author_dict:
            try:
                dept_name = author_dict[name_co][7].lower()
                col_name = author_dict[name_co][8].lower()

                if dept_name in dept_map:
                    dept_map[dept_name] = dept_map[dept_name] + 1
                else:
                    dept_map[dept_name] = 1

                if col_name in col_map:
                    col_map[col_name] =  col_map[col_name] + 1
                else:
                     col_map[col_name] = 1
            except:
                pass
    print(dept_map)
    print(col_map)
            
    for i in range(0, len(co_name)):
        a = co_name[i]
        if a in author_dict:

            #print(a)
            stat_a = get_stats(a, author_stats)
            #print(stat_a)
            #Get the dept and col names
            try:
                author_a_dept = author_dict[a][7].lower()
                author_a_col = author_dict[a][8].lower()
                
            except:
                continue
            
            if author_a_dept in dept_map:
                if dept_map[author_a_dept] >= 2:
                    stat_a[0] += 1
                    added = True
                    print(a)
            
            if author_a_col in col_map:
                if col_map[author_a_col] >= 2:
                    #print(col_name)
                    stat_a[1] += 1
                    added = True
                    #print(a)

            
            #if not added, then inc the uni count
            if len(dept_map) > 2 or len(col_map) > 2 or len(dept_map)+len(col_map)>2:
                #print(a)
                stat_a[2] += 1

            

            author_stats[a] = stat_a
    print("-------------")
     
```
